```python
import PyPDF2
import time
# creating a pdf file object
pdfFileObj = open('IDB Group Strategy with Panama 2021-2024 public_ENG.pdf', 'rb')
  
# creating a pdf reader object
pdfReader = PyPDF2.PdfReader(pdfFileObj)
```

```python

for n, page in enumerate(pdfReader.pages):
    print("=============",n,"==============" )
    page_content = page.extract_text()
    ask_bot_stream(page_content, False)
    time.sleep(1)
  
# closing the pdf file object
pdfFileObj.close()
```
